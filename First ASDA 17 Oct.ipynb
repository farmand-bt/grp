```python
group_5 = [
    'the', 'idea', 'of', 'decroissance', 'emerged', 'in', 'in', 'the', 'mid', 'as', 'a', 'direct', 'reaction', 'to', 'the', 'limits', 'to', 'growth', 'according', 'decroissance',
 'was', 'first', 'used', 'as', 'a', 'term', 'by', 'amar', 'in', 'a', 'rather', 'critically', 'reflect', 'upon', 'the', 'western', 'model', 'of', 'economic', 'development', 'translations',
    'of', 'some', 'of', 'the', 'works', 'of', 'georgescu', 'roegen', 'popularized', 'decroissance', 'as', 'an', 'alternative', 'to', 'the', 'already', 'established', 'zero', 'growth', 'loose', 'way', 'before', 'french', 'idea', 'of' 
]
word_freq = {}
for word in group_5:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

word_lengths = {word: len(word) for word in group_5}

combined_dict = {word: {'frequency': word_freq[word], 'length': word_lengths[word]} for word in group_5}
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(combined_dict, orient='index')
```
